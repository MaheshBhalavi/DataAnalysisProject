# Zomato Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the dataset

```python
zt=pd.read_csv('zomato.csv',encoding='latin1')
zt.head()
```

```python
zt.columns
```

```python
ztt=pd.read_excel('Country-Code.xlsx')
```

```python
ztt
```

# Combine two dataset

```python
zd=pd.merge(ztt,zt,on='Country Code',how='outer')
```

```python
zd
```

```python
zd.info()
```

```python
zd.describe()
```

```python
zd.isnull().sum()
```

```python
[features for features in zd.columns if zd[features].isnull().sum()>0] 
```

```python
zd['Country'].value_counts() #max number of zomato transactions happens in india 
```

```python
country_name=zd['Country'].value_counts().index
country_name
```

```python
country_val=zd['Country'].value_counts().values
country_val
```

```python
plt.figure(figsize=(15,8))
plt.pie(country_val[:5],labels=country_name[:5],autopct='%0.2f%%%%'); # top 5 country that zomato have  
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=zd,x='Country')
plt.title('Number of Restauranat in Each Country')
plt.xticks(rotation=90);
```

Observations:- Zomato major business is happens in India.
    Followed by USA,UK,Brazil and then South Africa. 

```python
Rating=zd.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_counts'})
Rating
```

```python
plt.figure(figsize=(15,8))
sns.barplot(data=Rating,x='Aggregate rating',y='Rating_counts');
```

```python
zd1=zd[zd['Aggregate rating']==0.0].groupby('Country').size().reset_index().rename(columns={0:'Count'})
zd1
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=zd1,x='Country',y='Count');
```

```python
zd3=zd.groupby(['City','Has Table booking']).size().reset_index().rename(columns={0:'Counts'})
zd3.head()
```

```python
zd4=zd3.loc[(zd3['Has Table booking']=='Yes')].rename(columns={0:'Counts'})
zd4
```

```python
zd5=zd3.loc[(zd3['Has Table booking']=='No')].rename(columns={0:'Counts'})
```

```python
plt.figure(figsize=(12,5))
zd3['Has Table booking'].value_counts().plot(kind='pie',autopct='%0.2f%%')
plt.title('Status of online booking');
```

```python
zd.shape
```

```python
plt.figure(figsize=(10,7))
zd4.groupby('City')['Counts'].max().nlargest(10).plot.bar()
plt.title('Top ten Citys in which online table booking facility is available');
```

```python
plt.figure(figsize=(10,7))
zd5.groupby('City')['Counts'].max().nlargest(10).plot.bar()
plt.title('Top ten Citys in which online table booking facility is not available');
```
